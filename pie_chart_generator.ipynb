```python
#           number of colors
#--------------------------------------
#            0   1    2    3    4    5
#--------------------------------------
# Bold 0    0%  11%  34%  33%  18%   4% 
#--------------------------------------
# Bold 2    0%   1%   9%  30%  40%  20% 
```

```python
def num_colors_distribution(deck, bold_tough, simulations):
    num_colors = []
    for i in range(simulations):
        np.random.shuffle(deck)
        flips = cl.generate_flipped_icons(deck,bold_tough)
        nc = sum( flips > 0 )
        num_colors.append(nc)
    return 100*plt.hist(num_colors,
                        np.arange(0,7)-0.5,
                        rwidth=0.95)[0]/simulations
```

```python
deck = cl.generate_deck(OO=3,O=11,OG=4,W=5,WG=2,OK=3,OB=3,KG=1,KK=2,G=2,BG=1,B=2,OKG=3)

num_colors_distribution(deck, 0, 10000)
```
